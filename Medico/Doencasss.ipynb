```python
df = pd.read_csv("Doen.csv")
df
```

```python
X = df.iloc[:,1:6].values
y = df.iloc[:,0].values
```

```python
from sklearn.preprocessing import LabelEncoder
encoder =  LabelEncoder()
y_convertido = encoder.fit_transform(y)
```

```python
y
```

```python
y = pd.get_dummies(y_convertido).values
```

```python
y
```

```python
sintoma1 = LabelEncoder()
X[:, 0] = sintoma1.fit_transform(X[:, 0])
X
```

```python
sintoma2 = LabelEncoder()
X[:, 1] = sintoma2.fit_transform(X[:, 1])
X
```

```python
sintoma3 = LabelEncoder()
X[:, 2] = sintoma3.fit_transform(X[:, 2])
X
```

```python
sintoma4 = LabelEncoder()
X[:, 3] = sintoma4.fit_transform(X[:, 3])
X
```

```python
remedio = LabelEncoder()
X[:, 4] = remedio.fit_transform(X[:, 4])
X
```

```python
X
```

```python
X = np.asarray(X).astype('float32')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='gelu'),
    tf.keras.layers.Dense(15, activation='gelu'),
    tf.keras.layers.Dense(5, activation='sigmoid')
  ])
model
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=50, epochs=100)
```

```python
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print('Resultado loss:', loss)
print('Resultado accuracy:', accuracy)
```
